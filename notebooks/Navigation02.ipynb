```python
def load_and_merge_ukhls_data(
    current_wave_path: str,
    previous_wave_path: str,
    variable_list: List[str],
    prev_target_vars: List[str],
    output_path: str
) -> pd.DataFrame:
    # 1. 加载当前波数据，去重变量
    df_curr_cols = ['pidp'] + variable_list
    df_curr_cols = list(dict.fromkeys(df_curr_cols))
    df_curr = pd.read_stata(current_wave_path, columns=df_curr_cols, convert_categoricals=False)

    # 2. 加载前一波目标变量数据
    df_prev = pd.read_stata(previous_wave_path, columns=['pidp'] + prev_target_vars, convert_categoricals=False)

    # 3. 重命名前一波变量为 xxx_prev（去掉 m_ 前缀）
    prev_renamed = {var: var.replace("m_", "") + "_prev" for var in prev_target_vars}
    df_prev.rename(columns=prev_renamed, inplace=True)

    # 4. 合并
    df_merged = df_curr.merge(df_prev, how='left', on='pidp')
    df_merged.to_csv(output_path, index=False)

    return df_merged
```
